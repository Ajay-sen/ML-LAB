**IRIS PLANT DATASET USING NAIVE BAYES**

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets

# metadata
print(iris.metadata)

# variable information
print(iris.variables)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.80)
```

**MULTINOMIAL Classifier**

```python
## MULTINOMIAL
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB().fit(X_train,y_train)
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
##Evaluation
from sklearn.metrics import classification_report,confusion_matrix
print("Confusion matrix")
print(confusion_matrix(y_test,y_pred))
print("------------------------------------")
print("------------------------------------")
print("Performance Evaluation")
print(classification_report(y_test,y_pred))
```

```python
## HYPERPARAMETER TUNING
alphas = np.linspace(0.01,10,25)
accuracies = []
```

```python
from sklearn.metrics import accuracy_score
for alpha in alphas:
    model = MultinomialNB(alpha=alpha)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
```

**Gaussian Classifier**

```python
##GAUSSIAN CLASSIFIER
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB().fit(X_train,y_train)
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
##Evaluation of the GuuassianNB
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print("-----------------------------------")
print("-----------------------------------")
print("Performance Evaluation")
print(classification_report(y_test,y_pred))
```

```python
## HYPERPARAMETER TUNING OF BERNOULLI CLASSIFIERS
var_smoothings = np.logspace(-20, 5, 30)
accuracies = []
```

```python
# Loop over var_smoothing values
for vs in var_smoothings:
    model = GaussianNB(var_smoothing=vs)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
```

```python
# Plot
plt.plot(var_smoothings, accuracies, marker='o')
plt.xscale('log')
plt.xlabel('var_smoothing')
plt.ylabel('Accuracy')
plt.title('GaussianNB on Iris: Accuracy vs var_smoothing')
plt.grid(True)
plt.show()
```

**Bernoulli Classifier**

```python
##BERNAULLI NB EVALUTION
from sklearn.naive_bayes import BernoulliNB
classifier=BernoulliNB().fit(X_train,y_train)
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
##Evaluation of Bernoulli Nb classifier
from sklearn.metrics import classification_report,confusion_matrix
print("Confusion matrix")
print(confusion_matrix(y_test,y_pred))
print("------------------------------------")
print("------------------------------------")
print("Performance Evaluation")
print(classification_report(y_test,y_pred))
```

```python
## HYPER PARAMETER TUNING
alphas = np.linspace(0, 2, 10)
binarizes = np.linspace(0.0, 1.0, 5)
results=[]
```

```python
for bin_thresh in binarizes:
    accuracies = []
    for alpha in alphas:
        model = BernoulliNB(alpha=alpha, binarize=bin_thresh, fit_prior=True, class_prior=None)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        accuracies.append(acc)
    results.append((bin_thresh, accuracies))
```

```python
plt.figure(figsize=(8,6))
for bin_thresh, accs in results:
    plt.plot(alphas, accs, marker='o', label=f'binarize={bin_thresh:.2f}')
plt.xlabel('Alpha')
plt.ylabel('Accuracy')
plt.title('BernoulliNB: Accuracy vs Alpha for different binarize thresholds')
plt.legend()
plt.grid(True)
plt.show()
```

**Observation: **

*When test size=0.6:*
  accuracies of Multinomial is 76 , Gaussian is 93 and Bernaulli is 31

*When test size=0.2:*
  accuracies of Multinomial is 90 , Gaussian is 97 and Bernaulli is 30
  
*When test size=0.8:*
  accuracies of Multinomial is 65 , Gaussian is 94 and Bernaulli is 33

```python
## Hyperparameter Tuning:
classifier=DecisionTreeClassifier(criterion='gini',max_depth=10)
classifier.fit(X_train,y_train)
```
